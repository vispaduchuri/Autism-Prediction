```python
import pandas as pd
import numpy as np
from sklearn import svm
import matplotlib.pyplot as plt
import seaborn as sns;sns.set(font_scale=1.2)
```

```python
cl=pd.read_csv('cleaned_demo3.csv')
```

```python
cl
```

```python
sns.lmplot('Class_ASD','result',data=cl,hue='Class_ASD',palette='Set1',fit_reg=False,scatter_kws={"s":50})
```

```python
age_res=cl[['jaundice','result']].as_matrix()
type_label=np.where(cl['Class_ASD']=='YES',0,1)
```

```python
model=svm.SVC(kernel='linear')
model.fit(age_res,type_label)
```

```python
#GET THE HYPERPLANES
w=model.coef_[0]
a=-w[0]/w[1]
xx=np.linspace(-1,2)
yy=a+xx-(model.intercept_[0]/w[1])
#PLOT THE PARALLELS TO THE SEPERATING HYPERPLANES
#THAT PASS THROUGH THE SUPPORT VECTORS 
b=model.support_vectors_[0]
yy_down=a+xx+(b[1]-a*b[0])
b=model.support_vectors_[-1]
yy_up=a+xx+(b[1]-a*b[0])
```

```python
#LOOK AT THE MARGINS AND SUPPORT VECTORS
sns.lmplot('jaundice','result',data=cl,hue='Class_ASD',palette='Set1',fit_reg=False,scatter_kws={"s":10})
plt.plot(xx,yy,linewidth=2,color='black')
plt.plot(xx,yy_down,'k--')
plt.plot(xx,yy_up,'k--')
plt.scatter(model.support_vectors_[:,0],model.support_vectors_[:,1],s=10,facecolors='none')
```

```python
sns.lmplot('jaundice','result',data=cl,hue='Class_ASD',palette='Set1',fit_reg=False,scatter_kws={"s":50})
plt.plot(xx,yy,linewidth=1,color='black')
```

```python
#enter the values for jaundice and result to check how the algorithm would classify it.
def jaun_res(jaundice,result):
    if(model.predict([[jaundice,result]]))==0 and jaundice==1:
        if(result<=10):
            print('is autistic')
        else:
            print("enter proper result")
    else:
        if(jaundice>=0 and jaundice<=1):
            if(result <7):
                            print('is not autistic')
            else:
                 print("autistic")
        else:
            print("enter proper values")
```

```python
jaun_res(0,6)
```
