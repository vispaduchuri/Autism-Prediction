```python
import pandas as pd
import numpy as np
from sklearn import svm
import matplotlib.pyplot as plt
import seaborn as sns;sns.set(font_scale=1.2)
```

```python
cl=pd.read_csv('ASD_3.csv')
```

```python
cl
```

```python
sns.lmplot('Res','cluster',data=cl,hue='cluster',palette='Set1',fit_reg=False,scatter_kws={"s":50})
```

```python
age_res=cl[['Jaundice','Res','BISE']].as_matrix()
type_label=np.where(cl['cluster']==0,0,1)
```

```python
model=svm.SVC(kernel='linear')
model.fit(age_res,type_label)
```

```python
#GET THE HYPERPLANES
w=model.coef_[0]
a=-w[0]/w[1]
xx=np.linspace(-1,2)
yy=a+xx-(model.intercept_[0]/w[1])
#PLOT THE PARALLELS TO THE SEPERATING HYPERPLANES
#THAT PASS THROUGH THE SUPPORT VECTORS 
b=model.support_vectors_[0]
yy_down=a+xx+(b[1]-a*b[0])
b=model.support_vectors_[-1]
yy_up=a+xx+(b[1]-a*b[0])
```

```python
#LOOK AT THE MARGINS AND SUPPORT VECTORS
sns.lmplot('Jaundice','Res',data=cl,hue='cluster',palette='Set1',fit_reg=False,scatter_kws={"s":10})
plt.plot(xx,yy,linewidth=2,color='black')
plt.plot(xx,yy_down,'k--')
plt.plot(xx,yy_up,'k--')
plt.scatter(model.support_vectors_[:,0],model.support_vectors_[:,1],s=10,facecolors='none')
```

```python
sns.lmplot('Jaundice','Res',data=cl,hue='cluster',palette='Set1',fit_reg=False,scatter_kws={"s":50})
plt.plot(xx,yy,linewidth=1,color='black')
```

```python
#enter the values for jaundice and result to check how the algorithm would classify it.
def jaun_res(Jaundice,Res,BISE):
    if(model.predict([[Jaundice,Res,BISE]]))==0 and Jaundice==1:
        if(Res>=8 and BISE<=0):
            print('is autistic')
        if(Res>6 and BISE<=0):
            print("Maybe Aspergers")
    else:
        if(Jaundice>=0 and Jaundice<=1):
            if(Res <7 and BISE>0 and BISE<3):
                            print('is not autistic')
            if(Res>=8 and BISE<=0):
                 print('is autistic')
        if(Res>6 and Res<8 and BISE<=0):
            print("Maybe Aspergers")
        if(Res<=6):
            print("is not autistic")
           
        
```
