```python
import pandas as pd
```

```python
lg=pd.read_csv("cleaned_demo4.csv")
```

```python
lg
```

```python
feature_cols=['age','result','jaundice','gender']
x=lg[feature_cols]
y=lg.Class_ASD
```

```python
#here we split the originial dataset into training set and testing set.
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

```python
#train the model by importing LOGISTICREGRESSION and feed the data.
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(x_train,y_train)
```

```python
y_pred_class=logreg.predict(x_test)
```

```python
from sklearn import metrics
print (metrics.accuracy_score(y_test,y_pred_class))
```

```python
confusion=metrics.confusion_matrix(y_test,y_pred_class)
TP=confusion[1,1]
TN=confusion[0,0]
FP=confusion[0,1]
FN=confusion[1,0]
```

```python
#shows how accurately the testing data is classified and show the precentage of the accuracy.
print (float(TP+TN)/float(TP+TN+FP+FN))
```
